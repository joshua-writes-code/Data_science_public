```python
#next visualizations random draws from the follwing distributions:
# 1) standard normal with mean of 0 and standard deviation of 0
# 2) normal with mean of 100 and standard deivation of 20
# 3) chi square with 0.5 degrees of freedom
# 4) Poisson with lambda of 1
# to sum up, it shows the results for the three continuos distributions and the discrete on (Poisson)
# the Poisson distribution is used for example to siomulate the arrival of rare external events
# like a jump in the price of an instrument or an exogenic shock.
# The following code generates the example of all of this.
# souce: python for fiance.

import numpy.random as npr
import matplotlib.pyplot as plt

sample_size = 500
rn1 = npr. standard_normal(sample_size)
rn2 = npr.normal(100, 20, sample_size)
rn3 = npr.chisquare(df=0.5, size=sample_size)
rn4 = npr.poisson(lam=1.0, size=sample_size)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
ax1.hist(rn1,bins=50)
ax1.set_title('standard normal', color= 'white')
ax1.set_ylabel('frequency', color= 'white')
ax2.hist(rn2, bins=50)
ax2.set_title('normal(100,20)', color= 'white')
ax3.hist(rn3, bins=25)
ax3.set_title('chi square', color= 'white')
ax3.set_ylabel('frequency', color= 'white')
ax4.hist(rn4, bins=25)
ax4.set_title('Poisson', color= 'white');
```
