```python
import numpy as np
import matplotlib.pyplot as plt
y = np.random.standard_normal((1000,2))

plt.figure(figsize=(10,6))
plt.plot(y[:,0], y[:,1], 'ro',color='black', marker='o')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('randomness by the machine', color='white')
```

```python
c = np.random.randint(0,10, len(y))
plt.figure(figsize=(10,6))
plt.scatter(y[:, 0], y[:,1], 
            c=c, 
            cmap='coolwarn',  
            marker='o')
plt.colorbar()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('randomness by the machine', color='white')
```

```python
fig, ax = plt.subplots(figsize=(10,6))
plt.boxplot(y)
plt.setp(ax, xticklabels=['x','y'])
plt.xlabel('data set', color='white')
plt.ylabel('value', color='white')
plt.title('boxplot',color='white')
```
